```python
import numpy as np
import pandas as pd
import category_encoders as ce
import os
from sklearn.model_selection import train_test_split
```

```python
curr_path = os.getcwd()
file_path = os.path.join(curr_path,'input/train.csv')
```

```python
df = pd.read_csv(file_path)
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,:-1]
```

```python
y = df.target
```

```python
X
```

```python
X.shape
```

```python
y
```

```python
df_train, df_val, ytrain, yval = train_test_split(X,y,test_size = 0.3,random_state=42)
```

```python
df_train.head()
```

```python
cols_to_enc = df_train.select_dtypes(object).columns.tolist()
print(cols_to_enc)
for c in df_train.columns:
    if c in cols_to_enc:
        print(c)
        tge = ce.TargetEncoder(cols=[c])
        tge.fit(pd.concat([df_train.loc[:,c],df_val.loc[:,c]],axis=0), pd.concat([ytrain,yval],axis=0))
        df_train.loc[:,c] = tge.transform(df_train.loc[:,c])
        df_val.loc[:,c] = tge.transform(df_val.loc[:,c])
```
